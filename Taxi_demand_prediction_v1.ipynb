**ЦЕЛЬ ПРОЕКТА:**

Для `сервиса такси` построить модель прогнозирования **количества заказов на следующий час**.

Это позволит эффективно управлять ресурсами и привлекать большее количество таксистов в пиковые часы.

Целевой KPI - `RMSE не выше 48` на тесте


**ЛИЧНЫЕ ЦЕЛИ:**

Научиться базовым принципам работы **с временными рядами** и определять их характеристики:
 - cтационарность 
 - сезонность
 - тренды
 
Испытать на практике возможности статистической автокорреляции `PCAF` и `ACF`, метод **дифференцирования**, а также познакомиться с возможностями бустера **XGBoost**.

```python
import pandas as pd
import numpy as np

#Графики
import matplotlib.pyplot as plt
import seaborn as sns

# Модели машинного обучения
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

import xgboost as xgb
from xgboost import plot_importance

# Метрики
from sklearn.metrics import mean_squared_error, mean_absolute_error

from sklearn.model_selection import train_test_split, validation_curve, learning_curve,   RandomizedSearchCV

# Преобразования
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# Пайплайны для кодирования и моделей
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.compose import ColumnTransformer

# Анализ тренда и сезонности
from statsmodels.tsa.seasonal import seasonal_decompose

# Тест Дики-Фулера для проверки стационарности ряда
from statsmodels.tsa.stattools import adfuller

# График для проверки автокорреляции
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# =======Функции в текущем проекте==================

# функция для отслеживания изменений во временном ряде
# после различных трансформаций
def print_time_series(df_time, samle_type_label, title):
    ''' Cтроит график временного ряда'''
    
    plt.figure(figsize=(10,5));


    (df_time
         # считаем скользящее среднее и стандартное отклонение 
         # для временного ряда
         .assign(
              rolling_mean = lambda x: x.num_orders.rolling(30).mean(),
              rolling_std = lambda x: x.num_orders.rolling(30).std()
         ) 
         .plot(ax=plt.gca())
    );

    plt.legend([samle_type_label, "Скользящее среднее", "Скользящее отклонение"], fontsize=12)
    plt.title(title, size=14, y=1.03)
    plt.ylabel("Количество заказов", size=12)
    plt.tick_params(axis='both', which='major', labelsize=14)
    plt.show()
   

def print_model_results(title, model, train_features, test_features, train_target, test_target):  
    
    '''обучает модель, выводит mae и rmse для обучающего и валидационного датасета'''
    
    model.fit(train_features, train_target)    
    print(title)  
    
    print("Обучающая выборка:    ", 
          f"RMSE: {mean_squared_error(model.predict(train_features), train_target) ** 0.5: .2f} |",
          f"MAE : {mean_absolute_error(model.predict(train_features), train_target): .2f}"
         )
    
    print("Валидационная выборка:", 
          f"RMSE: {mean_squared_error(model.predict(test_features), test_target) ** 0.5: .2f} |",
          f"MAE : {mean_absolute_error(model.predict(test_features), test_target): .2f}"
         ) 
    
    return model

#== функции для экспериментов с разным набором признаков

def test_models(train_dataset, validate_dataset):
    '''Обучает три базовые модели и выводит результаты
    возвращает обученные модели
    '''
    
    # обновляем модели, чтобы случайно не обучить их по 2 раза
    model_lr =  LinearRegression()
    model_rf =  RandomForestRegressor(n_estimators=20, random_state=1234)
    model_xgb = xgb.XGBRegressor(n_estimators=20, random_state=1234)

    for model, title in zip([model_xgb, model_lr, model_rf],
                        [">> XGBoost \n(n_estimators=20)\n", ">> Linear-Regression", ">> RandomForest \n(n_estimators=20)\n"]):
  
        print_model_results(title, model, 
                        train_dataset.drop(['num_orders'], axis=1),  validate_dataset.drop(['num_orders'], axis=1), 
                        train_dataset['num_orders'], validate_dataset['num_orders'])
        print() 
        
    return model_lr, model_rf, model_xgb
        
def test_linear_model(train_data, validate_data):     
    '''Обучает линейную регрессию и выводит результаты
    возвращает обученную модель
    '''
    
    model_lr = LinearRegression()

    print_model_results(">> Линейная Регрессия", model_lr, 
                        train_data.drop(['num_orders'], axis=1),  validate_data.drop(['num_orders'], axis=1), 
                        train_data['num_orders'], validate_data['num_orders'])
    
    return model_lr
```

### Upsampling данных 

**Аггрегируем** наши данные по часам и вычислим `суммарное` количество заказов. 

```python
# сохраняем результат ресемплинга в новом датафрейме
df_one_hour = df.resample("1H").sum()
display(df_one_hour.head(5))

print("Количество данных в семплированном датасете:", df_one_hour.shape[0])
```

- Датасет для предсказаний **сократился** `в 6 раз`!

```python
plt.figure(figsize=(10,5));

df_one_hour.plot(ax=plt.gca());

plt.legend(["Интервал 1 час"], fontsize=12)
plt.title("Динамика изменения количества заказов за 6 месяцев", size=14, y=1.03)
plt.ylabel("Количество заказов", size=12)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

- Картинка не сильно изменилась, но именно с ней нам предстоит работать при построении моделей


- Попробуем взять `upsampling` по **дням** и добавим **скользящие окна**  - каждая точка станет средним предыдущих `5 и 24 дней`

```python
plt.figure(figsize=(15,7));

# ресемплинг исходного датасета по доням
(df.resample("1D").sum()
     .assign(rolling_5 = lambda x: x.num_orders.rolling(5).mean())
     .assign(rolling_10 = lambda x: x.num_orders.rolling(24).mean())
     .plot(ax=plt.gca())
);

plt.legend(["Интервал 1 день", "Сглаживание 5 дней", "Сглаживание 24 дня"], fontsize=14)
plt.title("Динамика изменения количества заказов за 6 месяцев", size=16, y=1.03)
plt.ylabel("Количество заказов", size=12)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

- Теперь можно увидеть **четкий тренд**, как растет спрос на такси!

#### Функции для генерации признаков

```python
def discover_lag_features(df, lags_list):
    '''Генерирует временные ряды с лагами из списка
    
    сохраняет их в новые столбцы в датафрейме
    возвращает новый датафрейм
    '''
    
    data = df.copy()
    for lag in lags_list:
        data['lag_{}_hour'.format(lag)] = data['num_orders'].shift(lag)
    
    
    return data
```

```python
def discover_time_features(df):
    '''Генерирует календарные признаки 
    
    сохраняет их в новые столбцы в датафрейме
    возвращает новый датафрейм
    '''
    data = (df
              .copy()
           
               # генерируем календарные признаки
              .assign( 
                  hour       = lambda x: x.index.hour,
                  dayofweek  = lambda x: x.index.dayofweek,
                  dayofmonth = lambda x: x.index.day                 
              ) 
    )
    
    return data 

def discover_rolling_features(df, rolling_window_list):
    '''Генерирует "скользящие" признаки
    
    сохраняет их в новые столбцы в датафрейме
    возвращает новый датафрейм
    '''    
    data = df.copy()    
   
    # признаки - "скользящие" параметры случайной величины временного ряда
    # min, max, mean и std
    for window in rolling_window_list:
        data['rolling_mean_{}'.format(window)] = data['lag_1_hour'].rolling(window).mean()
        data['rolling_std_{}'.format(window)] = data['lag_1_hour'].rolling(window).std()
        data['rolling_max_{}'.format(window)] = data['lag_1_hour'].rolling(window).max()
        data['rolling_min_{}'.format(window)] = data['lag_1_hour'].rolling(window).min()
    
    return data
```

### Автокорреляция PACF - поиск закономерностей

В отличии от автокорреляции, `PACF` показывает не корреляцию `временного ряда` с самим собой, а **текущего наблюдения** в данных и его **лагом во времени**.

- Таким образом, PCAF не учитывает тренды и сезонность, а находит **корреляции**, которые может найти в `Residuals` временного ряда. 


- Отсюда и название `Partial Autocorrelation` Function - частичная автокорреляция.

Поэтому график PACF может помочь нам найти полезные **лаг-features** для моделирования!

```python
plt.figure(figsize=(10, 5))

plot_pacf(train_df.num_orders, ax=plt.gca(), method='ywm');

plt.title("PACF корреляция - данные с семплированием 1 час)", size = 14)
plt.xlabel("Лаги - 1 час назад, 2 часа назад и т.д.", size=12)
plt.ylabel("Корреляция", size=12)
plt.ylim([-0.5, 1.1])
plt.grid(True)
plt.show()
```

- `0-й лаг` - это текущее количество заказов, поэтому здесь `корреляция = 1`


- Достаточно много значений вошли в **95%-й доверительный интервал**, но мы возьмем только самые выдающиеся:)

### Предсказания на LAG-признаках

```python
# создаем временные датасеты для разных экспериментов
# для них генерируем разные наборы time-признаков
train = []
validate = []

# добавляем в обучающий и валидационный датасет сегенерированные признаки
train = discover_lag_features(train_cv_df, lags_list = range(1, 27))
validate = discover_lag_features(test_cv_df, lags_list = range(1, 27))

train.head(4)
```

#### Поиск лучших признаков c помощью XGBoost

```python
def learn_xgboost(estimators_count, features):
    '''функция для тестов
    обучает XGBoost на разных наборах фич
    '''
    model_xgb = xgb.XGBRegressor(n_estimators=estimators_count, random_state=123)
    model_xgb.fit(train[features], train['num_orders'],
             eval_set=[(train[features], train['num_orders']), 
                       (validate[features], validate['num_orders'])],
              
             # индикатор останова при проверке результатов на валидации
             # метрика должна улучшиться хотя бы в последних 30 итерациях
             early_stopping_rounds=30,
             verbose=True) 
    
    return model_xgb
```

```python
# Обучение XGBoost на датасете с 27 лагами
print("Валидационные результаты XGBoost - 27 временных лагов")
model_xgb = learn_xgboost(estimators_count=20, features=train.columns.drop(['num_orders']).to_list())
```

#### Оценка FeatureImportance с помощью LinearRegression

Посмотрим, как выглядят `веса признаков` для нашей Линейной модели.

```python
# сразу обернем в функцию
def create_features_dataframe(linear_model, features):
    '''возвращает датафрейм с весами линейной модели'''
    
    data = (pd.DataFrame({'features': features, 'importance': linear_model.coef_})
                .sort_values(by='importance', ascending=False)                
            )
    return data

# выводим таблицу с весами признаков с помощью '.coef_'
create_features_dataframe(model_lr, features_list).style.background_gradient(cmap='RdPu')
```

- `hour` и `dayofweek` заняли почетное место в нашем рейтинге и вошли в ТОП.


-  При этом результаты **Линейной Регрессии**  с новыми календарными признаками `не изменились`


-  Все новые `признаки оставляем` и сразу переходим к следующему этапу экспериментов.

```python
# обернем в функцию все предыдущие шаги с преобразованиями
def add_lag_time_features(df):   
    data = (df
                 .copy()
                 .pipe(discover_lag_features, best_lags)
                 .dropna()
                 .pipe(discover_time_features)  
                 # удаляем аутлайеры
                 .query('num_orders < 250')                
     )          
    return data
```
